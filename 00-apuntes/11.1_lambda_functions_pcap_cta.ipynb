# Funciones lambda en Python (PCAP)

Este notebook resume los puntos más importantes sobre **funciones lambda** para la preparación del **PCAP**.

## 1. Qué es una función lambda

Una **función lambda** es una función pequeña y anónima que se define en una sola línea.

Su forma general es:

```python
lambda parámetros: expresión
```

Características importantes:

- No suele llevar nombre directamente.
- Puede recibir cero, uno o varios parámetros.
- Solo puede contener **una expresión**.
- El valor de esa expresión es el valor que devuelve.

````python
sumar = lambda a, b: a + b

print(sumar(3, 5))
````

## 2. Sintaxis de una lambda

La sintaxis es:

```python
lambda arg1, arg2, ...: expresión
```

No se usa `return` porque el resultado de la expresión se devuelve automáticamente.

````python
cuadrado = lambda x: x ** 2
saludo = lambda nombre: "Hola " + nombre

print(cuadrado(4))
print(saludo("Ana"))
````

## 3. Diferencia entre `lambda` y `def`

Con `def` se define una función normal.  
Con `lambda` se define una función breve, normalmente para usarla en un contexto puntual.

### Con `def`

````python
def multiplicar(a, b):
    return a * b

print(multiplicar(2, 6))
````

### Con `lambda`

````python
multiplicar_lambda = lambda a, b: a * b

print(multiplicar_lambda(2, 6))
````

En general:

- `def` se usa para funciones más legibles o complejas.
- `lambda` se usa para funciones cortas y simples.

## 4. Una lambda solo puede tener una expresión

Este es uno de los puntos más importantes para el PCAP.

Una lambda:

- **sí** puede hacer cálculos o evaluaciones en una expresión.
- **no** puede contener varias instrucciones, bucles completos o varias sentencias separadas.

````python
# Válido
par_o_impar = lambda x: "par" if x % 2 == 0 else "impar"
print(par_o_impar(7))
print(par_o_impar(10))
````

Lo siguiente no sería válido en una lambda:

```python
# No válido
# lambda x:
#     y = x + 1
#     return y
```

Para eso debe usarse `def`.

## 5. Parámetros en funciones lambda

Las lambdas pueden tener distintos tipos de parámetros.

````python
sin_parametros = lambda: "Python"
uno = lambda x: x * 10
varios = lambda a, b, c: a + b + c

print(sin_parametros())
print(uno(7))
print(varios(1, 2, 3))
````

## 6. Lambda como función inline

Una lambda se usa muchas veces en el mismo sitio donde hace falta, sin guardarla previamente en una variable.

````python
print((lambda x, y: x - y)(10, 3))
````

## 7. Funciones de orden superior

Una **función de orden superior** es una función que:

- recibe otra función como argumento, o
- devuelve una función.

Las lambdas se usan mucho en este contexto.

````python
def aplicar(funcion, valor):
    return funcion(valor)

# podriamos guardar una lista de lambda. Es un tipo en concreto lambda
lista_lambdas= [lambda x: x*2, lambda x: x * 3]
print(lista_lambdas[0](5))

resultado = aplicar(lambda x: x * 3, 5)
print(resultado)
````

## 8. Uso con `map()`

`map()` aplica una función a cada elemento de un iterable.

Sintaxis:

```python
map(función, iterable)
```

````python
numeros = [2, 2, 3, 4]
dobles_list = list(map(lambda x: x * 2, numeros))
dobles_set = set(map(lambda x: x * 2, numeros))
print(type(map(lambda x: x * 2, numeros)))

print(numeros)
print(dobles_list)
print(dobles_set)
````

## 9. Uso con `filter()`

`filter()` selecciona los elementos para los que una función devuelve `True`.

````python
numeros = [1, 2, 3, 4, 5, 6]
pares = list(filter(lambda x: x % 2 == 0, numeros))

print(pares)
````

## 10. Uso con `sorted()` y el parámetro `key`

`sorted()` puede ordenar usando una función de clave.

Aquí `lambda` es muy útil.

````python
palabras = ["python", "sol", "programacion", "ia"]
ordenadas_por_longitud = sorted(palabras, key=lambda palabra: len(palabra), reverse = True)

print(ordenadas_por_longitud)
````

También se puede ordenar una lista de tuplas o diccionarios.

````python
personas = [ # Lista de tuplas
    ("Luis", 25),
    ("Ana", 19),
    ("Marta", 31)
]

ordenadas_por_edad = sorted(personas, key=lambda persona: persona[0])
print(ordenadas_por_edad)
````

## 11. Uso con `min()` y `max()` usando `key`

También pueden usar una función para decidir el criterio de comparación.

````python
palabras = ["casa", "extraordinario", "sol", "ordenador"]

mas_corta = min(palabras, key=lambda p: len(p))
mas_larga = max(palabras, key=lambda p: len(p))

print("Más corta:", mas_corta)
print("Más larga:", mas_larga)
````

## 12. Casos típicos donde sí merece la pena usar lambda

Lambda suele ser útil cuando:

- la operación es muy corta,
- se usa una sola vez,
- se pasa como argumento a otra función.

````python
notas = [("Ana", 8.5), ("Luis", 6.2), ("Eva", 9.1)]

ordenadas = sorted(notas, key=lambda x: x[1], reverse=True)
print(ordenadas)
````

## 13. Cuándo no conviene usar lambda

No conviene usar lambda cuando:

- la lógica es larga,
- la expresión resulta difícil de leer,
- hacen falta varias instrucciones,
- interesa documentar mejor la función.

En esos casos es preferible `def`.

````python
def clasificar_numero(x):
    if x > 0:
        return "positivo"
    elif x < 0:
        return "negativo"
    return "cero"

print(clasificar_numero(5))
print(clasificar_numero(-2))
print(clasificar_numero(0))
````

## 14. Errores o ideas clave para el PCAP

Conviene recordar:

- `lambda` crea funciones anónimas.
- Solo puede contener **una expresión**.
- Devuelve automáticamente el resultado de esa expresión.
- Se usa mucho con `map()`, `filter()`, `sorted()`, `min()` y `max()`.
- Para funciones complejas, es mejor `def`.

## 15. Resumen final

Las funciones lambda son una forma breve de definir funciones pequeñas.  
Son muy útiles cuando se necesita una operación corta y rápida, sobre todo al trabajar con otras funciones que reciben funciones como argumento.

Para PCAP, lo más importante es dominar:

- la sintaxis,
- la diferencia con `def`,
- la restricción de una sola expresión,
- el uso con `map`, `filter`, `sorted`, `min` y `max`.

# Closures en Python (PCAP)

Este notebook recoge los conceptos esenciales sobre **closures** para la preparación del **PCAP**.  
Se explican de forma sencilla, con ejemplos claros y progresivos.

## 1. Qué es una función anidada

Una **función anidada** es una función definida dentro de otra función.

```python
def externa():
    def interna():
        print("Estoy dentro de la función interna")
    interna()

externa()
```

Las closures parten precisamente de este tipo de funciones.

## 2. Qué es un closure

Un **closure** aparece cuando:

- una función está definida dentro de otra,
- la función interna utiliza variables de la función externa,
- la función externa termina,
- y la función interna sigue conservando acceso a esas variables.

Es decir, la función interna "recuerda" el entorno en el que fue creada.

```python
def externa():
    mensaje = "Hola desde el closure"

    def interna():
        print(mensaje)

    return interna

f = externa()
f()
```

Aunque `externa()` ya terminó, la función `interna()` sigue teniendo acceso a `mensaje`.

## 3. Variables libres

Una **variable libre** es una variable usada dentro de una función pero definida en un ámbito exterior.

En este ejemplo, `mensaje` es una variable libre para `interna()`.

```python
def crear_saludo():
    mensaje = "Buenos días"

    def saludar():
        return mensaje

    return saludar

f = crear_saludo()
print(f())
```

## 4. Captura del ámbito exterior

Una closure captura valores del entorno donde fue creada.

```python
def multiplicador(n):
    def multiplicar(x):
        return x * n
    return multiplicar

por_2 = multiplicador(2)
por_5 = multiplicador(5)

print(por_2(10))
print(por_5(10))
```

Cada función devuelta recuerda su propio valor de `n`.

## 5. Persistencia del estado

Uno de los puntos más importantes es que una closure puede mantener estado entre llamadas.

```python
def contador():
    valor = 0

    def incrementar():
        nonlocal valor
        valor += 1
        return valor

    return incrementar

c = contador()

print(c())
print(c())
print(c())
```

Aquí `valor` no se pierde entre llamadas a `incrementar()`.

## 6. Uso de `nonlocal`

La palabra clave `nonlocal` permite modificar una variable del ámbito exterior más cercano que no sea global.

Sin `nonlocal`, Python intentaría tratar esa variable como local dentro de la función interna.

```python
def prueba():
    x = 10

    def interna():
        nonlocal x
        x += 1
        return x

    return interna

f = prueba()
print(f())
print(f())
```

## 7. Qué pasa si no se usa `nonlocal`

Si se intenta modificar una variable exterior sin `nonlocal`, aparece error.

```python
def ejemplo_error():
    x = 5

    def interna():
        # x += 1   # Esto produciría un error si se descomentara
        return x

    return interna

f = ejemplo_error()
print(f())
```

Si solo se lee la variable, no hace falta `nonlocal`.  
Si se quiere **modificar**, sí hace falta.

```python
x = "global"

def externa():
    y = "nonlocal"

    def interna():
        z = "local"
        #y += " modificado"
        nonlocal y
        y += " modificado"
        #x += " modificado"
        global x
        x += " modificado"
        print("z =", z)
        print("y =", y)
        print("x =", x)

    interna()

print(x)
externa()
print(x)
```

## 8. Closure frente a función normal

Una función normal simplemente ejecuta código.  
Una closure, además, conserva información del entorno donde se creó.

```python
def funcion_normal(x):
    return x + 1

print(funcion_normal(10))
```

```python
def crear_sumador(n):
    def sumar(x):
        return x + n
    return sumar

sumar_10 = crear_sumador(10)
print(sumar_10(7))
```

## 9. Closure como generadora de funciones

Este es un uso muy típico: crear funciones personalizadas a partir de otra función.

```python
def crear_potencia(exponente):
    def elevar(base):
        return base ** exponente
    return elevar

cuadrado = crear_potencia(2)
cubo = crear_potencia(3)

print(cuadrado(4))
print(cubo(4))
```

## 10. Encapsular datos con closures

Las closures permiten ocultar variables internas sin necesidad de crear una clase.

```python
def crear_cuenta(saldo_inicial):
    saldo = saldo_inicial

    def operar(cantidad):
        nonlocal saldo
        saldo += cantidad
        return saldo

    return operar

cuenta = crear_cuenta(100)
print(cuenta(50))
print(cuenta(-20))
print(cuenta(10))
```

La variable `saldo` queda encapsulada en la closure.

## 12. Aplicación práctica simple

Un closure puede servir para personalizar comportamientos.

```python
def crear_prefijo(prefijo):
    def unir(texto):
        return prefijo + texto
    return unir

agregar_info = crear_prefijo("[INFO] ")
agregar_error = crear_prefijo("[ERROR] ")

print(agregar_info("Proceso iniciado"))
print(agregar_error("Fallo en la conexión"))
```

## 13. Diferencia entre closure y variable global

Con una global, la función depende de un valor externo general del programa.  
Con una closure, la función conserva su propio contexto.

```python
mensaje_global = "Hola global"

def usar_global():
    return mensaje_global

print(usar_global())
```

```python
def crear_mensaje(texto):
    def mostrar():
        return texto
    return mostrar

m1 = crear_mensaje("Hola A")
m2 = crear_mensaje("Hola B")

print(m1())
print(m2())
```

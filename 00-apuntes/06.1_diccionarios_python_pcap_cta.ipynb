# Diccionarios en Python para PCAP

## Objetivo
Este notebook explica de forma clara y ordenada los conceptos más importantes de los **diccionarios** en Python, con enfoque práctico para la preparación del **PCAP**.

Un **diccionario** es una estructura de datos que permite guardar información en formato **clave–valor**.  
Cada clave identifica un dato, y cada valor es la información asociada a esa clave.

Ejemplo general:

```python
alumno = {
    "nombre": "Ana",
    "edad": 19,
    "aprobado": True
}
```

En este ejemplo:

- `"nombre"`, `"edad"` y `"aprobado"` son las **claves**
- `"Ana"`, `19` y `True` son los **valores**


## 1. Crear diccionarios

Un diccionario se crea usando llaves `{}` y separando cada par **clave: valor** con comas.

También se puede crear un diccionario vacío y rellenarlo después.


````python
# Diccionario con datos
persona = {
    "nombre": "Luis",
    "edad": 30,
    "ciudad": "Madrid"
}

print(persona)

# Diccionario vacío
vacio = {}
print(vacio)
````

## 2. Estructura clave–valor

La idea fundamental de un diccionario es que cada dato se guarda asociado a una clave.

Esto permite acceder a la información por su nombre lógico, no por su posición, como ocurre en las listas.

### Diferencia conceptual
- En una **lista**, se accede por índice: `lista[0]`
- En un **diccionario**, se accede por clave: `diccionario["nombre"]`


````python
producto = {
    "id": 101,
    "nombre": "Teclado",
    "precio": 29.99
}

print(producto["id"])
print(producto["nombre"])
print(producto["precio"])
````

## 3. Acceso a valores mediante clave

Para obtener el valor asociado a una clave se utiliza la sintaxis:

```python
diccionario[clave]
```

Esto funciona bien si la clave existe.  
Si la clave no existe, Python genera un error `KeyError`.


````python
coche = {
    "marca": "Toyota",
    "modelo": "Corolla",
    "km": 85000
}

print(coche["marca"])
print(coche["km"])

# Descomentar la línea siguiente produciría KeyError
# print(coche["color"])
````

````python
class Persona:
    pass
lista_diccionarios = [{"334A":Persona("Juan",23,"334A")},{"558B":Persona("Marta",46,"558B")}]

print(lista_diccionarios[1],["555B"].edad)
# 1-Se accede al 1 diccionario de la lista, 2 - Cortchete a la edad del objeto persona dentro del diccionario con etiqueta "558B"
````

## 4. Modificación de valores existentes

Si la clave ya existe, al asignarle un nuevo valor se **actualiza** el dato almacenado.


````python
coche = {
    "marca": "Toyota",
    "modelo": "Corolla",
    "km": 85000
}

print("Antes:", coche)

coche["km"] = 90000

print("Después:", coche)
````

## 5. Añadir nuevos pares clave–valor

Si la clave no existía anteriormente, Python la crea al hacer la asignación.


````python
coche = {
    "marca": "Toyota",
    "modelo": "Corolla"
}

coche["color"] = "Blanco"
coche["puertas"] = 5

print(coche)
````

## 6. Eliminar elementos del diccionario

Hay varias formas de eliminar elementos:

- `del diccionario[clave]` elimina una clave concreta.
- `pop(clave)` elimina la clave y devuelve su valor.
- `clear()` elimina todos los elementos del diccionario.


````python
datos = {
    "nombre": "Marta",
    "edad": 22,
    "curso": "Python"
}

print("Original:", datos)

del datos["curso"]
print("Tras del:", datos)

edad = datos.pop("edad")
print("Valor eliminado con pop:", edad)
print("Tras pop:", datos)

datos.clear()
print("Tras clear:", datos)
````

## 7. Comprobar si una clave existe

Para comprobar si una clave está en un diccionario se usa el operador `in`.

Esto es importante para evitar errores cuando no se sabe si una clave existe.


````python
usuario = {
    "login": "admin",
    "activo": True
}

print("login" in usuario) # Comprobar si clave esta en el diccionario.
print("email" in usuario)
print("email" not in usuario)
````

## 8. Recorrer diccionarios

Al recorrer un diccionario con un `for`, por defecto se recorren las **claves**.


````python
persona = {
    "nombre": "Elena",
    "edad": 27,
    "ciudad": "Sevilla"
}

for clave in persona: 
    print(clave)
````

## 9. Recorrer claves, valores e ítems

Python permite recorrer:

- solo las **claves**
- solo los **valores**
- los pares **clave–valor**


````python
persona = {
    "nombre": "Elena",
    "edad": 27,
    "ciudad": "Sevilla"
}

print("Claves:")
for clave in persona:
    print(clave)

print("\nValores:")
for valor in persona.values():
    print(valor)

print("\nClaves y valores:")
for clave, valor in persona.items():
    print(clave, "->", valor)
````

## 10. Uso de `keys()`, `values()` e `items()`

Estos métodos son fundamentales:

- `keys()` devuelve las claves
- `values()` devuelve los valores
- `items()` devuelve pares `(clave, valor)`

No devuelven listas normales, sino objetos especiales iterables.


````python
alumno = {
    "nombre": "Raúl",
    "edad": 20,
    "nota": 8.7
}

print(alumno.keys()) # Listado con todas las claves del diccionario. 
print(alumno.values()) # Listado con los valores. 
print(alumno.items()) # Tuplas clave,valor del diccionario.

# Si se necesita una lista explícita
print(list(alumno.keys()))
print(list(alumno.values()))
print(list(alumno.items()))
````

## 11. Uso de `get()`

El método `get()` permite obtener el valor de una clave sin provocar error si no existe.

Sintaxis:

```python
diccionario.get(clave)
diccionario.get(clave, valor_por_defecto)
```


````python
config = {
    "modo": "oscuro",
    "idioma": "es"
}

print(config.get("modo"))
print(config.get("fuente"))
print(config.get("fuente", "Arial"))
````

## 12. Diferencia entre `[]` y `get()`

### Con corchetes `[]`
- devuelve el valor si la clave existe
- produce `KeyError` si no existe

### Con `get()`
- devuelve el valor si la clave existe
- devuelve `None` o un valor por defecto si no existe

Esta diferencia es muy importante en el examen y en la práctica real.


````python
datos = {"a": 10, "b": 20}

print(datos["a"])
print(datos.get("a"))

# print(datos["z"])   # KeyError
print(datos.get("z"))          # None
print(datos.get("z", 0))       # 0
````

## 13. Longitud del diccionario con `len()`

La función `len()` devuelve el número de pares clave–valor que contiene el diccionario.


````python
inventario = {
    "raton": 15,
    "teclado": 8,
    "monitor": 4
}

print(len(inventario))
````

## 14. Diccionarios mutables

Los diccionarios son **mutables**, lo que significa que pueden cambiar después de ser creados.

Se puede:
- modificar valores
- añadir nuevas claves
- eliminar elementos

Esto los diferencia de tipos inmutables como `str`, `int` o `tuple`.


````python
d = {"x": 1, "y": 2}
print("Original:", d)

d["x"] = 100
d["z"] = 300
del d["y"]

print("Modificado:", d)
````

## 15. Claves únicas

En un diccionario no puede haber dos claves iguales.

Si se repite una clave, la última asignación sobrescribe a la anterior.


````python
d = {
    "nombre": "Carlos",
    "nombre": "Lucía",
    "edad": 25
}

print(d)
````

## 16. Restricciones sobre las claves: tipos hashables

Las claves de un diccionario deben ser de tipos **inmutables** o, más precisamente, **hashables**.

### Ejemplos válidos como clave
- `str`
- `int`
- `float`
- `bool`
- `tuple` (si contiene elementos inmutables)

### Ejemplos no válidos como clave
- `list`
- `dict`
- `set`

Esto ocurre porque Python necesita identificar cada clave de forma estable.


````python
d = {
    "uno": 1,
    2: "dos",
    (1, 2): "tupla"
}

print(d)

# Ejemplo no válido:
# invalido = {
#     [1, 2]: "lista"
# }
````

## 17. Diccionarios anidados

Un diccionario puede contener otros diccionarios como valores.

Esto permite representar estructuras más complejas.


````python
empresa = {
    "nombre": "TechSoft",
    "direccion": {
        "calle": "Gran Vía",
        "numero": 15,
        "ciudad": "Madrid"
    },
    "empleados": 120
}

print(empresa)
print(empresa["direccion"])
print(empresa["direccion"]["ciudad"])
````

## 18. Conversión entre diccionario y otras colecciones

A veces interesa transformar un diccionario o parte de él en lista, tupla u otras estructuras.

Ejemplos frecuentes:
- convertir claves a lista
- convertir valores a lista
- convertir ítems a lista de tuplas


````python
capitales = {
    "España": "Madrid",
    "Francia": "París",
    "Italia": "Roma"
}

print(list(capitales))
print(list(capitales.keys()))
print(list(capitales.values()))
print(list(capitales.items()))
````

## 19. Métodos útiles adicionales

Aunque los puntos anteriores son los más importantes, conviene conocer algunos métodos que suelen aparecer al trabajar con diccionarios.

### `update()`
Permite añadir o actualizar varios elementos a la vez.

### `popitem()`
Elimina y devuelve el último par clave–valor insertado.

### `copy()`
Crea una copia superficial del diccionario.


````python
d1 = {"a": 1, "b": 2}
d1.update({"b": 20, "c": 30})
print("Tras update:", d1)

ultimo = d1.popitem()
print("Elemento eliminado con popitem:", ultimo)
print("Diccionario final:", d1)

d2 = d1.copy()
print("Copia:", d2)
````

## 20. Errores frecuentes con diccionarios

### Error 1: usar una clave que no existe
```python
d["clave"]
```
si no existe, genera `KeyError`.

### Error 2: pensar que se recorren valores por defecto
Al hacer:
```python
for x in d:
```
se recorren **claves**, no valores.

### Error 3: usar tipos mutables como claves
Una lista no puede ser clave.

### Error 4: olvidar que las claves son únicas
Si se repite una clave, se pierde el valor anterior.

Los diccionarios son una de las estructuras de datos más importantes en Python y aparecen con frecuencia en el **PCAP**.

### Ideas clave que deben quedar claras

- Se accede a los valores usando la **clave**
- Las **claves son únicas**
- Los diccionarios son **mutables**
- `get()` evita errores cuando no se sabe si una clave existe
- `keys()`, `values()` e `items()` son esenciales para recorrerlos

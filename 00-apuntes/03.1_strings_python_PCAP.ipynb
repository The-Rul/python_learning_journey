## 1) Qué es `str` y la idea de **inmutabilidad** <a id="que-es-str-e-inmutabilidad"></a>

Un **string** en Python es un objeto de tipo `str` que representa texto (Unicode).

### Inmutabilidad
`str` es un tipo **inmutable**:
- Una vez creado, **no se puede modificar** carácter a carácter.
- Operaciones que “parecen modificar” un string en realidad **crean uno nuevo**.

Esto explica por qué no se puede hacer `s[0] = 'X'` y por qué muchos métodos devuelven una nueva cadena.

[Ir al índice](#contenidos)

````python
s = "hola"
print(type(s))

# Intento de modificación directa (no permitido)
try:
    s[0] = "H"
except TypeError as e:
    print("Error:", e)

# En su lugar, se crea una nueva cadena
s2 = "H" + s[1:]
print("Original:", s)
print("Nueva:", s2)
````

## 2) Creación de strings: comillas simples/dobles y multilínea <a id="creacion-de-strings"></a>

Se pueden crear strings con:
- Comillas simples: `'texto'`
- Comillas dobles: `"texto"`
- Triple comilla (multilínea): `'''...'''` o `"""..."""`

La triple comilla conserva saltos de línea y es útil para textos largos.

[Ir al índice](#contenidos)

````python
a = 'Hola "Mundo"'
b = "Hola Mundo"
c = """Línea 1
Línea 2
Línea 3"""

print(a, b)
print(c)
````

## 3) Caracteres de escape <a id="caracteres-de-escape"></a>

Dentro de una cadena se pueden usar secuencias de escape con `\`:

- `\n` salto de línea
- `\t` tabulación
- `\\` barra invertida
- `\'` comilla simple
- `\"` comilla doble

[Ir al índice](#contenidos)

````python
texto = "Primera línea\nSegunda línea\t(con tabulación)"
print(texto)

ruta_windows = "C:\\Users\\Jose\\Desktop"
print(ruta_windows)

comillas = "Dijo: \"hola\" y se fue."
print(comillas)
````

## 4) Raw strings (`r"..."`) <a id="raw-strings"></a>

Un **raw string** se escribe anteponiendo `r` o `R`:

```python
r"texto"
```

En un raw string, Python **no interpreta** (o interpreta mínimamente) secuencias de escape como `\n` o `\t`.  
Es muy útil para rutas de Windows y expresiones regulares (regex).

> Detalle práctico: un raw string **no puede terminar** con una sola barra invertida `\`.

[Ir al índice](#contenidos)

````python
normal = "C:\\temp\\nuevo"
raw = r"C:\temp\nuevo" #Uso en rutas para no coger los n y t

print("Normal:", normal)
print("Raw   :", raw)

print("Normal interpreta \\n ->", "A\nB")
print("Raw no interpreta \\n ->", r"A\nB")
````

## 5) Indexación <a id="indexacion"></a>

- Índices comienzan en `0`
- Índices negativos cuentan desde el final (`-1` es el último)

[Ir al índice](#contenidos)

````python
s = "Python"
print(s[0])   # 'P'
print(s[3])   # 'h'
print(s[-1])  # 'n'
print(s[-2])  # 'o'
````

## 6) Slicing <a id="slicing"></a>

```python
s[inicio:fin:paso]
```

- `inicio` incluido, `fin` excluido
- `paso` opcional
- `s[::-1]` invierte el string

[Ir al índice](#contenidos)

````python
s = "programación"
print(s[0:4])     # 'prog'
print(s[:4])      # 'prog'
print(s[4:])      # desde índice 4
print(s[::2])     # saltando de 2 en 2
print(s[::-1])    # toda palabra pero saltando -1 (invertido)
print(s[-3:-1])    # toda palabra pero saltando -1 (invertido)

````

## 7) Concatenación y repetición <a id="concatenacion-y-repeticion"></a>

- `+` concatena
- `*` repite

[Ir al índice](#contenidos)

````python
a = "Hola"
b = "PCAP"
print(a + " " + b)

print("ha" * 3)
print("-" * 10)
````

## 8) Longitud, pertenencia e iteración <a id="longitud-pertenencia-iteracion"></a>

- `len(s)` longitud
- `'sub' in s` pertenencia
- `for ch in s` iteración

[Ir al índice](#contenidos)

````python
s = "certificación"
print(len(s))

print("ción" in s)
print("xyz" not in s)
print("xyz" not in s)


for ch in "abc":
    print(ch)
````

## 9) Comparación de strings (orden lexicográfico / Unicode) <a id="comparacion-de-strings"></a>

Los strings se comparan **lexicográficamente** (carácter a carácter) usando Unicode.  
Es sensible a mayúsculas/minúsculas.

[Ir al índice](#contenidos)

````python
print("Ana" < "Bea")
print("casa" < "caso")
print("a" < "B")      # depende de Unicode
print("hola" == "hola")
print("hola" == "Hola")
````

## 10) `ord()` y `chr()`, `min()` y `max()` <a id="funciones-ord-chr-min-max"></a>

- `ord(c)` → código Unicode
- `chr(n)` → carácter
- `min(s)` / `max(s)` → por orden Unicode

[Ir al índice](#contenidos)

````python
print(ord("A"), ord("a"))
print(chr(65), chr(97))

s = "Zebra"
print(min(s), max(s))
````

## 11) Transformación: `lower()`, `upper()`, etc. <a id="metodos-de-transformacion"></a>

- `lower()`, `upper()`
- `capitalize()`, `title()`
- `swapcase()`

Devuelven **un nuevo string**.

[Ir al índice](#contenidos)

````python
s = "pCaP y PYTHON"
print("Lower: ",s.lower())
print("Upper: ",s.upper())
print("Capitalize: ",s.capitalize())
print("Title: ",s.title())
print("Swapcase: ",s.swapcase())
print("Original:", s)
````

## 12) Limpieza: `strip()`, `lstrip()`, `rstrip()` <a id="limpieza-strip"></a>

Eliminan espacios (u otros caracteres indicados) en los extremos.

[Ir al índice](#contenidos)

````python
s = "   hola   "
print(s.strip())
print(s.lstrip())
print(s.rstrip())

t = "...texto...."
print(t.strip("."))

u = "__--dato--__"
print(u.strip("_-"))

u = "__--da--to--__"
print(u.strip("_-"))
````

## 13) Búsqueda y conteo <a id="busqueda-y-conteo"></a>

- `find()` devuelve índice o `-1`
- `rfind()` busca desde el final
- `index()` devuelve índice o lanza `ValueError`
- `count()` cuenta ocurrencias
- `startswith()` / `endswith()`

[Ir al índice](#contenidos)

````python
s = "banana"

print(s.find("na"))
print(s.find("xyz"))

try:
    print(s.index("xyz"))
except ValueError as e:
    print("Error:", e)

print(s.count("na"))
print(s.startswith("ba"))
print(s.endswith("na"))
````

## 14) Reemplazo: `replace()`  <a id="reemplazo"></a>

`replace(viejo, nuevo[, count])` devuelve un nuevo string.

[Ir al índice](#contenidos)

````python
s = "2026-03-03"
print(s.replace("-", "/"))
print(s.replace("-", "/", 1))
````

## 15) `split()`, `splitlines()` y `join()` <a id="division-y-union"></a>

- `split()` divide en lista
- `splitlines()` separa por líneas
- `join()` une partes con un separador

[Ir al índice](#contenidos)

````python
s = "uno dos   tres"
print(s.split())

csv = "a,b,c"
print(csv.split(","))

texto = "L1\nL2\nL3"
print(texto.splitlines())

partes = ["2026", "03", "03"]
print("-".join(partes))
````

## 16) Tests de contenido <a id="tests-de-contenido"></a>

- `isalpha()`, `isdigit()`, `isalnum()`, `isspace()`
- `islower()`, `isupper()`, `istitle()`

[Ir al índice](#contenidos)

````python
print("abc".isalpha())
print("123".isdigit())
print("abc123".isalnum())
print("   ".isspace())

print("hola".islower())
print("HOLA".isupper())
print("Hola Mundo".istitle())
````

## 17) Conversión y validación: `str()`, `int()`, `float()` <a id="conversion-y-validacion"></a>

Conversión desde string puede fallar con `ValueError` si el formato es inválido.

[Ir al índice](#contenidos)

````python
x = 42
print(str(x), type(str(x)))

print(int("123"))
print(float("1.5"))

try:
    print(int("12a"))
except ValueError as e:
    print("Error:", e)
````

## 18) Formateo: f-strings y `format()` <a id="formateo"></a>

- f-strings: `f"Hola {nombre}"`
- `format()`: `"Hola {}".format(nombre)`

Formato numérico típico: `:.2f`, `:x`, etc.

[Ir al índice](#contenidos)

````python
nombre = "Ana"
edad = 20
nota = 8.756

print(f"Hola {nombre}, edad={edad}")
print(f"Nota: {nota:.2f}")

n = 255
print(f"Decimal {n} -> hex {n:x}")

print("Hola {}, edad={}".format(nombre, edad))
````

## 19) `encode()` / `decode()` (Unicode y bytes) <a id="unicode-y-codificacion"></a>

- `encode()` convierte `str` → `bytes`
- `decode()` convierte `bytes` → `str`

Codificación típica: `utf-8`.

`UTF-8` es un sistema de codificación de caracteres (Unicode) que convierte letras, símbolos y emojis en bytes para poder guardarlos y transmitirlos. Es el estándar más usado porque es compatible con ASCII y puede representar prácticamente cualquier idioma.

[Ir al índice](#contenidos)

````python
s = "canción"
b = s.encode("utf-8")
print(b, type(b))

s2 = b.decode("utf-8")
print(s2, type(s2))

# Decodificar con otra codificación puede producir texto raro (mojibake) o errores según el caso.
print("latin-1:", b.decode("latin-1"))
````

# Programación Orientada a Objetos (POO) en Python


## ¿Qué es la orientación a objetos?

En Python, un **objeto** agrupa **datos (estado)** y **funciones (comportamiento)** en una misma entidad.

- **Clase**: “molde” o plantilla.
- **Instancia/objeto**: un ejemplar creado a partir de la clase.



## 1) Definir una clase y crear objetos
Una clase se define con `class`.


```python
class Persona:
    pass

p1 = Persona()
p2 = Persona()

print(p1)
print(p2)
print("¿Son el mismo objeto?", p1 is p2)
```

## 2) `__init__`: el inicializador
`__init__` se ejecuta **justo después** de crear el objeto. Sirve para **inicializar** sus atributos.

> Nota: `__init__` no “crea” el objeto (eso lo hace `__new__` internamente). `__init__` solo lo configura.


```python
class Producto:
    def __init__(self, nombre, precio):
        self.nombre = nombre
        self.precio = precio

p = Producto("Teclado", 29.99)
print(p.nombre, p.precio)
```

### Errores típicos con `__init__`
- Olvidar `self`.
- No asignar atributos con `self.`.


```python
class Mal:
    def __init__(self, x):
        # Esto NO crea un atributo de instancia; es solo una variable local.
        x = x

m = Mal(10)
print("¿Tiene atributo x?", hasattr(m, "x")) # !! "hasattr" metodo importante para verificar si un objeto tiene un atributo específico
```

## 3) ¿Qué es `self`?
`self` es una referencia al **objeto actual** (la instancia) dentro de los métodos.

- El nombre `self` es una **convención** (podría llamarse distinto, pero no se recomienda).
- Python pasa la instancia automáticamente al llamar a un método de instancia.


```python
class Contador:
    def __init__(self):
        self.valor = 0

    def incrementar(self): # Metodo de instancia.Llama a self, la instancia sobre la que se llaa el método.
        # self apunta a la instancia sobre la que se llama el método
        self.valor += 1

c = Contador()
c.incrementar()
c.incrementar()
print("valor:", c.valor)

# Equivalente "manual": llamar al método pasando la instancia explícitamente. 
Contador.incrementar(c) # LLamar al metodo de clase pasando la instancia como argumento.
print("valor tras llamada manual:", c.valor)
```

## 4) Variables de instancia
Las **variables de instancia** (atributos) viven en cada objeto.

- Se suelen crear en `__init__` con `self.atributo = ...`.
- Cada instancia tiene sus propios valores.


```python
class CuentaBancaria:
    def __init__(self, titular, saldo=0): # Saldo tiene valor por defecto 0
        self.titular = titular
        self.saldo = saldo

c1 = CuentaBancaria("Ana", 100)
c2 = CuentaBancaria("Luis", 50)

c1.saldo += 25  # solo cambia c1
print(c1.titular, c1.saldo)
print(c2.titular, c2.saldo)
```

### Añadir atributos “al vuelo” (dinámicamente)
Python permite añadir atributos en cualquier momento, aunque no siempre es buena práctica.


```python
class Punto:
    def __init__(self, x, y):
        self.x = x
        self.y = y

pt = Punto(2, 3)
pt.z = 99  # atributo añadido dinámicamente
print(pt.x, pt.y, pt.z)
print("Atributos en __dict__:", pt.__dict__) # Devuelve diccionario con los attributos del onjeto.
```

## 5) Métodos de instancia
Un **método de instancia** es una función definida dentro de una clase que recibe `self`.

Se usa para:
- Modificar atributos (`self...`).
- Consultar el estado.
- Encapsular comportamiento.


```python
class Rectangulo:
    def __init__(self, ancho, alto):
        self.ancho = ancho
        self.alto = alto

    def area(self):
        return self.ancho * self.alto

    def escalar(self, factor):
        self.ancho *= factor
        self.alto *= factor

r = Rectangulo(4, 2)
print("Área:", r.area())
r.escalar(3)
print("Nuevo ancho/alto:", r.ancho, r.alto)
print("Nueva área:", r.area())
```

### `self` vs. variables locales
Dentro de un método, una variable sin `self.` es local y desaparece al terminar.


```python
class DemoLocalVsInstancia:
    def __init__(self):
        self.total = 0

    def sumar(self, n):
        total = 100  # variable local, NO es atributo
        self.total += n
        return total

d = DemoLocalVsInstancia()
print("retorno:", d.sumar(5))
print("atributo total:", d.total)
print("¿Tiene atributo 'total' local?", hasattr(d, "total"))  # sí, por self.total
print("¿Tiene atributo 'total' local de 100?", d.total == 100)
```

## 6) “Variables privadas” en Python
Python no tiene privacidad estricta como Java/C++.

Hay **convenciones**:
- `nombre`: público
- `_nombre`: “protegido” por convención (uso interno)
- `__nombre`: *name mangling* (se “renombra” internamente para evitar colisiones)

### 6.1) Convención con un guion bajo `_`


```python
class Servicio:
    def __init__(self):
        self._token = "ABC123"  # convención: uso interno

s = Servicio()
print(s._token)  # se puede acceder, pero se considera interno
```

### 6.2) Doble guion bajo `__` y *name mangling*
Cuando un atributo empieza por `__` (y no termina también con `__`), Python lo transforma internamente a `_Clase__atributo`.

Esto **no es seguridad**, pero ayuda a evitar conflictos en herencia.


```python
class CajaFuerte:
    def __init__(self, clave):
        self.__clave = clave  # name mangling

    def comprobar(self, intento):
        return intento == self.__clave

cf = CajaFuerte("1234")

print("¿Existe __clave?", hasattr(cf, "__clave"))
print("Contenido __dict__:", cf.__dict__)

# Acceso 'mangling' (no recomendado salvo debugging)
print("Acceso con name mangling:", cf._CajaFuerte__clave)

print("Comprobar 0000:", cf.comprobar("0000"))
print("Comprobar 1234:", cf.comprobar("1234"))
```

### 6.3) Herencia y por qué existe el *name mangling*
Si una clase hija usa un `__atributo` con el mismo nombre, **no pisa** el de la clase padre.


```python
class Padre:
    def __init__(self):
        self.__x = "padre"

    def ver(self):
        return self.__x

class Hijo(Padre):
    def __init__(self):
        super().__init__()
        self.__x = "hijo"

h = Hijo()
print("ver():", h.ver())
print("__dict__:", h.__dict__)
print("Padre __x:", h._Padre__x)
print("Hijo __x:", h._Hijo__x)
```

## 7) `hasattr`, `getattr`, `setattr`, `delattr`
Estas funciones permiten **inspección y acceso dinámico**.

- `hasattr(obj, 'attr')` → True/False
- `getattr(obj, 'attr', default)` → valor o `default`
- `setattr(obj, 'attr', value)` → crea/modifica
- `delattr(obj, 'attr')` → elimina


```python
class Config:
    def __init__(self):
        self.modo = "dev"

cfg = Config()

print("hasattr(modo):", hasattr(cfg, "modo"))
print("getattr(modo):", getattr(cfg, "modo"))
print("getattr(inexistente, 'N/A'):", getattr(cfg, "inexistente", "N/A"))

setattr(cfg, "debug", True)
print("debug:", cfg.debug)

delattr(cfg, "modo")
print("hasattr(modo) tras borrado:", hasattr(cfg, "modo"))
```

### Ejemplo práctico: validar si un objeto cumple una “interfaz” mínima
Sin usar typing avanzado, se puede comprobar si existen métodos requeridos.


```python
class ExportadorCSV:
    def exportar(self, datos):
        return "csv..."

class ExportadorJSON:
    def exportar(self, datos):
        return "json..."

class Roto:
    def guardar(self, datos):
        return "oops"

def usar_exportador(obj, datos):
    if not hasattr(obj, "exportar"):
        raise TypeError("El objeto no tiene el método requerido: exportar()")
    return obj.exportar(datos)

print(usar_exportador(ExportadorCSV(), [1,2,3]))
print(usar_exportador(ExportadorJSON(), [1,2,3]))

try:
    usar_exportador(Roto(), [1,2,3])
except TypeError as e:
    print("ERROR:", e)
```

## 8) `__dict__`: el diccionario interno de atributos
En la mayoría de objetos “normales”, los atributos de instancia se guardan en `obj.__dict__`.

- Es muy útil para depurar.
- También permite modificar atributos, pero hay que tener cuidado.


```python
class Libro:
    def __init__(self, titulo, paginas):
        self.titulo = titulo
        self.paginas = paginas

l = Libro("Dune", 800)
print(l.__dict__)

# Modificación directa (posible, pero mejor usar atributos o métodos)
l.__dict__["paginas"] = 900
print("paginas:", l.paginas)

# Añadir atributos por __dict__
l.__dict__["autor"] = "Frank Herbert"
print(l.autor)
print(l.__dict__)
```

### ¿Siempre existe `__dict__`?
No siempre. Por ejemplo, si una clase define `__slots__`, puede no existir `__dict__`.


```python
class Ligero:
    __slots__ = ("x", "y")  # evita __dict__ por defecto

    def __init__(self, x, y):
        self.x = x
        self.y = y

obj = Ligero(1, 2)
print(obj.x, obj.y)
print("¿Tiene __dict__?", hasattr(obj, "__dict__"))

try:
    obj.z = 3
except AttributeError as e:
    print("No se puede crear 'z' porque __slots__ lo impide:", e)
```

## 9) Destrucción de objetos: `del`, `__del__` y “destroy”
En Python, los objetos se destruyen cuando **no quedan referencias** a ellos (recolección de basura).

### 9.1) `del`
- `del variable` elimina la **referencia**, no “borra” mágicamente el objeto si hay otras referencias.


```python
class Nodo:
    def __init__(self, valor):
        self.valor = valor

n1 = Nodo(10)
n2 = n1  # otra referencia al mismo objeto

print("Antes:", n1.valor, n2.valor)
del n1  # elimina solo la referencia n1

# n2 sigue apuntando al objeto
print("Después:", n2.valor)
```

### 9.2) `__del__` (destructor)
Puedes definir `__del__` para ejecutar código cuando el objeto se destruye.

**OJO:**
- No hay garantía de cuándo se ejecutará.
- En programas con ciclos de referencia o en algunos intérpretes/escenarios, puede ejecutarse tarde o nunca.
- No se recomienda para liberar recursos críticos; para eso se usan **context managers** (`with`).


```python
import gc

class Recurso:
    def __init__(self, nombre):
        self.nombre = nombre
        print(f"[__init__] Recurso creado: {self.nombre}")

    def __del__(self):
        print(f"[__del__] Recurso destruido: {self.nombre}")

r = Recurso("archivo-temporal")

# Forzamos eliminación de la referencia
del r

# Podemos pedir al GC que intente recolectar (en CPython suele ser inmediato)
gc.collect() # Forzar recolección de basura para ver el efecto de __del__ (no garantizado en otros intérpretes)
print("Fin de la celda.")
```

### 9.3) Patrón “destroy/close”: liberar recursos de forma explícita
En Python es común implementar un método como `close()` o `destroy()` para cerrar recursos.

Ejemplo típico: un objeto que abre un archivo o una conexión.


```python
class Sesion:
    def __init__(self):
        self.activa = True
        print("Sesión abierta")

    def destroy(self):
        if self.activa:
            self.activa = False
            print("Sesión cerrada (destroy)")
        else:
            print("Sesión ya estaba cerrada")

s = Sesion()
s.destroy()
s.destroy()
```

### 9.4) Mejor práctica: context manager (`with`)
Si el objeto gestiona recursos, lo ideal es implementar `__enter__` y `__exit__`.


```python
class SesionCM:
    def __init__(self):
        self.activa = False

    def __enter__(self):
        self.activa = True
        print("SesiónCM abierta")
        return self

    def __exit__(self, exc_type, exc, tb):
        self.activa = False
        print("SesiónCM cerrada automáticamente")
        # False => si hubo excepción, no la “silencia”
        return False

with SesionCM() as ses: # Se llama al método __enter__ al entrar al bloque, y se asigna el resultado a 'ses'. 
    print("¿Activa dentro?", ses.activa)

print("¿Activa fuera?", ses.activa) # Al salir del bloque, se llama a __exit__ automáticamente.
```

## 10) Mini-proyecto final: clase completa con todo lo visto
Una clase que combina:
- `__init__`
- `self`
- atributos de instancia
- métodos de instancia
- atributo “privado” con `__`
- inspección con `hasattr` y `__dict__`
- método `destroy`


```python
class Usuario:
    def __init__(self, nombre, email):
        self.nombre = nombre
        self.email = email
        self.__password = None      # “privado” (name mangling)
        self.activo = True

    def set_password(self, pwd):
        # ejemplo simple (NO es un hash real)
        self.__password = f"hash({pwd})"

    def check_password(self, intento):
        return self.__password == f"hash({intento})"

    def destroy(self):
        # “cierre” lógico del objeto
        self.activo = False
        self.__password = None

u = Usuario("Marta", "marta@example.com")
u.set_password("secreto")

print("¿Tiene nombre?", hasattr(u, "nombre"))
print("¿Tiene __password?", hasattr(u, "__password"))
print("Atributos (__dict__):", u.__dict__)
print("Login correcto:", u.check_password("secreto"))
print("Login incorrecto:", u.check_password("otro"))

u.destroy()
print("Activo:", u.activo)
print("Password tras destroy:", u._Usuario__password)
print("Estado final __dict__:", u.__dict__)
```

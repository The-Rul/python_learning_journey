## 1. Recordatorio rápido: listas

- Son **ordenadas**.
- Son **mutables**.
- Permiten **duplicados**.
- Se escriben con corchetes: `[]`.

Ejemplo:


```python
lista = [10, 20, 20, 30]
print(lista)
print(type(lista))
```

A partir de esta referencia será más fácil entender las diferencias con **tuplas** y **sets**.


# 2. Tuplas

## 2.1 Qué es una tupla

Una **tupla** es una estructura de datos que puede almacenar varios valores en una sola variable.

Sus características principales son:

- Es **ordenada**.
- Es **inmutable**.
- Permite **duplicados**.
- Se suele escribir con paréntesis: `()`.

### Diferencia clave con las listas

La diferencia más importante es esta:

- La **lista** se puede modificar.
- La **tupla** no se puede modificar una vez creada.


```python
tupla = (10, 20, 20, 30)
print(tupla)
print(type(tupla))
```

## 2.2 Creación de tuplas

Se pueden crear de varias formas:


```python
tupla1 = (1, 2, 3)
tupla2 = 4, 5, 6
tupla_vacia = ()

print(tupla1)
print(tupla2)
print(tupla_vacia)
```

## 2.3 Tupla de un solo elemento

Este es un punto muy importante en PCAP.

Una tupla de un solo elemento **necesita una coma**.


```python
no_es_tupla = (5)
si_es_tupla = (5,)

print(no_es_tupla, type(no_es_tupla))
print(si_es_tupla, type(si_es_tupla))
```

### Diferencia con listas

En listas no existe este problema:

- `[5]` siempre es una lista con un elemento.
- `(5)` no es una tupla; es solo un entero entre paréntesis.


## 2.4 Indexación y slicing en tuplas

Como las listas, las tuplas son **ordenadas**, por lo que se puede acceder a sus elementos por índice y usar slicing.


```python
tupla = ('a', 'b', 'c', 'd', 'e')

print(tupla[0])
print(tupla[-1])
print(tupla[1:4])
```

### Diferencia con sets

Esto sí se puede hacer con **listas** y **tuplas**, pero **no** con sets, porque los sets no son ordenados.


## 2.5 Inmutabilidad de las tuplas

Una vez creada una tupla, no se pueden cambiar, añadir ni eliminar elementos.

Esto contrasta directamente con las listas.


```python
lista = [1, 2, 3]
tupla = (1, 2, 3)

lista[0] = 100
print('Lista modificada:', lista)

try:
    tupla[0] = 100
except TypeError as e:
    print('Error al modificar la tupla:', e)
```

### Idea importante para examen

Que una tupla sea inmutable significa que **su estructura no cambia**. Esto la hace útil cuando se quiere guardar información que no debe modificarse accidentalmente.


## 2.6 Operadores con tuplas

Las tuplas permiten algunas operaciones típicas parecidas a las listas, como concatenación y repetición.


```python
t1 = (1, 2)
t2 = (3, 4)

print(t1 + t2)
print(t1 * 3)
print(2 in t1)
print(9 not in t1)
```

### Diferencia con listas

Estas operaciones también existen en listas, pero la diferencia sigue siendo que la lista luego sí puede modificarse y la tupla no.


## 2.7 Funciones y métodos útiles

Las tuplas tienen pocos métodos, precisamente porque son inmutables.

Los dos más importantes son:

- `count()` → cuenta cuántas veces aparece un valor.
- `index()` → devuelve la posición de la primera aparición.


```python
tupla = (10, 20, 20, 30, 20)

print(tupla.count(20))
print(tupla.index(30))
print(len(tupla))
```

## 2.8 Empaquetado y desempaquetado

Otro punto muy importante en Python y en PCAP es el uso de tuplas para **packing** y **unpacking**.


```python
# packing
datos = 'Ana', 19, 'Madrid'
print(datos)

# unpacking
nombre, edad, ciudad = datos
print(nombre)
print(edad)
print(ciudad)
```

### Diferencia con listas

Aunque también se puede desempaquetar una lista, este uso está muy asociado a las tuplas, sobre todo cuando una función devuelve varios valores.


## 2.9 Tuplas como retorno de funciones

En Python es muy común que una función devuelva varios valores. En la práctica, esos valores se devuelven empaquetados en una tupla.


```python
def division_y_resto(a, b):
    return a // b, a % b

resultado = division_y_resto(17, 5)
print(resultado) # Resultado es una tupla

cociente, resto = division_y_resto(17, 5) # Con unpacking cada uno de los datos en var diferente
print('Cociente:', cociente)
print('Resto:', resto)
```

## 2.10 Conversión entre listas y tuplas


```python
lista = [1, 2, 3]
tupla = tuple(lista)
lista2 = list(tupla)

print(tupla)
print(lista2)
```

### Idea práctica

A veces se convierte una tupla a lista para poder modificar los datos y luego volver a convertirla en tupla.


# 3. Sets

## 3.1 Qué es un set

Un **set** es una colección de elementos:

- **no ordenada**
- **mutable**
- sin elementos duplicados

Se suele usar cuando interesa trabajar con valores únicos.


```python
conjunto = {1, 2, 3, 3, 4, 4, 5}
print(conjunto)
print(type(conjunto))
```

### Diferencia con listas

- La lista mantiene el orden de inserción.
- El set no se usa como estructura indexada.
- La lista permite duplicados.
- El set elimina duplicados automáticamente.


## 3.2 Crear sets correctamente

Hay que recordar una cosa muy importante:

- `{}` crea un **diccionario vacío**, no un set vacío.
- Para crear un set vacío se usa `set()`.


```python
diccionario_vacio = {}
set_vacio = set()

print(type(diccionario_vacio))
print(type(set_vacio))
```

## 3.3 Añadir y eliminar elementos

Como los sets son mutables, se pueden modificar.


```python
numeros = {1, 2, 3}

numeros.add(4)
print(numeros)

numeros.remove(2)
print(numeros)

numeros.discard(10)
print(numeros)
```

### Diferencia entre remove() y discard()

- `remove(x)` da error si `x` no existe.
- `discard(x)` no da error si `x` no existe.


```python
numeros = {1, 2, 3}

try:
    numeros.remove(99)
except KeyError as e:
    print('Error con remove:', e)

numeros.discard(99)
print('discard no produce error')
```

## 3.4 Uso de pop(), clear() y copy()


```python
conjunto = {10, 20, 30}
copia = conjunto.copy()
print('Copia:', copia)

eliminado = conjunto.pop()
print('Elemento eliminado con pop():', eliminado)
print('Set después de pop():', conjunto)

conjunto.clear()
print('Set después de clear():', conjunto)
```

### Punto importante

En un set, `pop()` elimina un elemento **arbitrario**, no el “último”, porque el set no es una estructura ordenada como una lista.


## 3.5 Operaciones con sets

Este es uno de los bloques más importantes.

Los sets se usan mucho para operaciones de teoría de conjuntos.


```python
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

print('Unión:', a | b)
print('Intersección:', a & b)
print('Diferencia a-b:', a - b)
print('Diferencia simétrica:', a ^ b)
```

También existen los métodos equivalentes:

- `union()`
- `intersection()`
- `difference()`
- `symmetric_difference()`


```python
a = {1, 2, 3}
b = {3, 4, 5}

print(a.union(b))
print(a.intersection(b))
print(a.difference(b))
print(a.symmetric_difference(b))
```

### Diferencia con listas

Las listas no tienen este comportamiento de teoría de conjuntos de forma natural.

Si se quieren eliminar duplicados o calcular intersecciones de manera cómoda, suele ser mejor usar sets.


## 3.6 Comprobaciones entre conjuntos


```python
a = {1, 2}
b = {1, 2, 3, 4}
c = {7, 8}

print(a.issubset(b))
print(b.issuperset(a))
print(a.isdisjoint(c))
```

Estas operaciones son muy típicas de examen cuando se trabaja con sets.


## 3.7 Iteración y pertenencia

Se puede recorrer un set con `for` y comprobar si un elemento pertenece al conjunto con `in`.


```python
conjunto = {'rojo', 'verde', 'azul'}

for color in conjunto:
    print(color)

print('rojo' in conjunto)
print('negro' not in conjunto)
```

### Diferencia con listas y tuplas

En los tres casos se puede usar `in`, pero en listas y tuplas también se puede acceder por índice. En sets no.


## 3.8 Conversión entre listas y sets

Una aplicación muy frecuente es eliminar duplicados de una lista.


```python
lista = [1, 2, 2, 3, 4, 4, 5]
sin_duplicados = list(set(lista))

print('Lista original:', lista)
print('Lista sin duplicados:', sin_duplicados)
```

### Advertencia importante

Este proceso elimina duplicados, pero puede alterar el orden. Por eso no siempre sirve si el orden importa.


## 3.9 frozenset

`frozenset` es una versión inmutable de un set.

Es decir:

- `set` → mutable
- `frozenset` → inmutable


```python
fs = frozenset([1, 2, 3])
print(fs)
print(type(fs))

try:
    fs.add(4)
except AttributeError as e:
    print('Error:', e)
```

# 4. Comparación directa: listas vs tuplas vs sets


```python
comparacion = {
    'Lista': {
        'Ordenada': True,
        'Mutable': True,
        'Permite duplicados': True,
        'Indexación': True,
        'Sintaxis habitual': '[]'
    },
    'Tupla': {
        'Ordenada': True,
        'Mutable': False,
        'Permite duplicados': True,
        'Indexación': True,
        'Sintaxis habitual': '()'
    },
    'Set': {
        'Ordenada': False,
        'Mutable': True,
        'Permite duplicados': False,
        'Indexación': False,
        'Sintaxis habitual': '{,,,} o set()'
    }
}

for estructura, datos in comparacion.items():
    print(f'--- {estructura} ---')
    for clave, valor in datos.items():
        print(f'{clave}: {valor}')
    print()
```

# 5. Cuándo usar cada una

## Lista
- Cuando se necesita una colección ordenada y modificable.
- Cuando importa la posición de los elementos.

## Tupla
- Cuando los datos no deben cambiar.
- Cuando se quiere devolver varios valores desde una función.
- Cuando interesa una estructura más fija.

## Set
- Cuando se quieren valores únicos.
- Cuando interesa eliminar duplicados.
- Cuando se van a realizar operaciones de unión, intersección o diferencia.


# 6. Resumen final para PCAP

## Tuplas
- Son ordenadas e inmutables.
- Permiten duplicados.
- Admiten indexación y slicing.
- Tienen pocos métodos: `count()` e `index()`.
- Son muy usadas para packing y unpacking.

## Sets
- Son mutables pero no indexados.
- No permiten duplicados.
- Son ideales para operaciones de conjuntos.
- `set()` crea un set vacío; `{}` crea un diccionario vacío.
- `remove()` y `discard()` no se comportan igual.
- `pop()` elimina un elemento arbitrario.

## Diferencias con listas
- La lista es mutable y ordenada.
- La tupla es ordenada pero inmutable.
- El set elimina duplicados y no se usa por índice.

## Excepciones más comunes en Python

- **TypeError**  
  Se produce cuando se intenta realizar una operación con **tipos de datos incompatibles**.

- **ValueError**  
  Ocurre cuando una función recibe un **valor del tipo correcto pero con un contenido inválido**.

- **IndexError**  
  Se genera cuando se intenta acceder a un **índice que no existe en una secuencia** (listas, tuplas, etc.).

- **KeyError**  
  Aparece cuando se intenta acceder a una **clave que no existe en un diccionario**.

- **ZeroDivisionError**  
  Se produce cuando se intenta **dividir un número entre cero**.

- **AttributeError**  
  Ocurre cuando se intenta acceder a un **atributo o método que un objeto no tiene**.

- **ImportError**  
  Se genera cuando **Python no puede importar un módulo o un elemento de un módulo**.

- **FileNotFoundError**  
  Aparece cuando se intenta **abrir un archivo que no existe en la ruta indicada**.

- **AssertionError**  
  Se lanza cuando una **instrucción `assert` evalúa una condición como falsa**.

Las excepciones que más aparecen en ejercicios del examen son:
- TypeError
- ValueError
- IndexError
- KeyError
- ZeroDivisionError
- AssertionError

## 6. except

## 10. else
El bloque else en un try se ejecuta solo si el código dentro de try se ejecuta sin producir ninguna excepción. 
>Importante. Suele aparecer en ejercicios del examen PCAP.

````python
try:
  x = 10/2
except ZeroDivisionError:
  print('Error')
else:    print('Resultado:',x)
````

## 11. finally
El bloque finally se ejecuta siempre al terminar el try, haya ocurrido o no una excepción.

````python
try:
  x=10/0
except:
  print('Error')
finally:
  print('Siempre se ejecuta')
````

````python
try:
  x=10/0
finally:
  print('Siempre se ejecuta') #Se ejecuta y salta la excepción sin manejar
print('Esto no se ejecuta porque no se ha manejado la excepción')
````

## 16. Propagación de excepciones
La propagación de excepciones es el mecanismo mediante el cual una excepción que se produce en una parte del programa se transmite hacia arriba en la cadena de llamadas de funciones hasta encontrar un bloque except que la capture.

````python
def dividir():
    return 10 / 0

def calcular():
    return dividir()

try:
    calcular()
except ZeroDivisionError:
    print("Error: división entre cero")
````

````python
def dividir():
    return 10/0

def calcular():
    try: 
        resultado = dividir()
    except ZeroDivisionError as z:
        resultado = 0
        print("Error en calcular")
        raise z
    return resultado

try: 
    print(calcular())
except ZeroDivisionError:
    print("Error : division entre cero")
    
print("Finaliza")
````

## 17. Re-lanzar excepción
Relanzar una excepción significa volver a lanzar una excepción que ya ha sido capturada en un bloque except para que continúe propagándose por el programa.
Esto se hace cuando:
- se quiere registrar o procesar parcialmente el error
- pero no se quiere manejar completamente en ese punto
- por lo que se deja que otra parte del programa lo gestione

````python
try:
  x=10/0
except ZeroDivisionError:
  print('Error detectado')
  raise
````

## 18. assert
assert es una instrucción que se utiliza para comprobar que una condición es verdadera durante la ejecución del programa.

Si la condición es verdadera, el programa continúa normalmente.
Si la condición es falsa, Python genera una excepción: `AssertionError`

````python
x=-1
assert x>=0, 'x debe ser positivo'
````

Para qué se usa assert
assert se usa principalmente para:
- verificar suposiciones del programa
- detectar errores durante el desarrollo
- comprobar que ciertas condiciones internas se cumplen

Por ejemplo:
- validar estados internos de una función
- comprobar que un valor calculado es correcto
- detectar errores de lógica durante el desarrollo

````python
def dividir(a, b):
    assert b != 0, "El divisor no puede ser cero"
    return a / b
````

## Buenas prácticas al manejar excepciones en Python

Al trabajar con excepciones es importante seguir ciertas recomendaciones para que el código sea más claro, seguro y mantenible.

### 1. Capturar excepciones específicas
Evita capturar todas las excepciones con `except:`. Es mejor indicar el tipo concreto de error.

```python
try:
    numero = int(valor)
except ValueError:
    print("El valor introducido no es un número válido")
```
### 2. Evitar bloques `try` demasiado grandes
El bloque `try` debe contener únicamente el código que puede generar la excepción.

#### Incorrecto
```python
try:
    x = int(valor)
    resultado = 10 / x
    print(resultado)
except ValueError:
    print("Error")
```

#### Correcto
```python
try:
    x = int(valor)
except ValueError:
    print("Valor no válido")

resultado = 10 / x
```

### 3. Usar finally para liberar recursos

El bloque finally se ejecuta siempre y es útil para cerrar archivos o conexiones.
```python
try:
    archivo = open("datos.txt")
    contenido = archivo.read()
finally:
    archivo.close()
```

### 4. No usar excepciones para controlar la lógica normal del programa

Las excepciones deben utilizarse para situaciones excepcionales, no para el flujo normal.
### Incorrecto
```python
try:
    numero = int(texto)
except:
    numero = 0
```
### Correcto
```python
if texto.isdigit():
    numero = int(texto)
else:
    numero = 0
```

### 5. Registrar o mostrar información útil del error
```python
try:
    int("hola")
except ValueError as e:
    print("Ocurrió un error:", e)
```

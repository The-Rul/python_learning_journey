## 1. Qué es I/O en Python

**I/O** significa **Input/Output**, es decir, **entrada y salida** de datos.

En Python, esto suele referirse a dos situaciones muy habituales:

- **Entrada**: el programa recibe datos.
- **Salida**: el programa muestra o guarda datos.

### Ejemplos de entrada
- El usuario escribe algo por teclado.
- El programa lee información de un archivo.

### Ejemplos de salida
- El programa muestra información por pantalla.
- El programa escribe información en un archivo.

En PCAP, normalmente I/O se estudia en dos bloques:

1. **Entrada y salida por consola**
2. **Entrada y salida con archivos**


## 2. Entrada por teclado con `input()`

La función `input()` permite leer datos que escribe el usuario por teclado.


````python
nombre = input("Introduce tu nombre: ")
print("Hola,", nombre)
````

### Idea importante para PCAP

`input()` **siempre devuelve una cadena de texto (`str`)**, aunque el usuario escriba números.


````python
edad = input("Introduce tu edad: ")
print(edad)
print(type(edad))
````

Esto es una fuente muy común de errores en principiantes.

Por ejemplo, si el usuario escribe `20`, eso no se guarda como entero, sino como texto `"20"`.


## 3. Conversión de tipos después de `input()`

Como `input()` devuelve texto, muchas veces hay que convertir el valor a otro tipo.

Las conversiones más habituales son:

- `int()`
- `float()`
- `str()`


````python
edad = int(input("Introduce tu edad: "))
altura = float(input("Introduce tu altura en metros: "))

print("Edad:", edad, type(edad))
print("Altura:", altura, type(altura))
````

### Ejemplo de error típico


````python
texto = input("Escribe un número: ")

try:
    resultado = texto + 5
except TypeError as e:
    print("Error:", e)
````

Ese error ocurre porque `texto` es una cadena y `5` es un entero.

La solución correcta sería convertir antes:


````python
texto = input("Escribe un número: ")
numero = int(texto)
resultado = numero + 5

print("Resultado:", resultado)
````

## 4. Salida por pantalla con `print()`

La función `print()` se usa para mostrar información por pantalla.


````python
print("Hola")
print("Python")
print(123)
print(3.14)
print(True)
````

También puede mostrar varias cosas a la vez:


````python
nombre = "Ana"
edad = 19

print("Nombre:", nombre, "Edad:", edad)
````

## 5. Parámetros importantes de `print()`: `sep` y `end`

### `sep`
Sirve para indicar qué separador se coloca entre varios elementos que se imprimen.


````python
print("rojo", "verde", "azul")
print("rojo", "verde", "azul", sep=" - ")
print("2026", "03", "17", sep="/")
````

### `end`
Sirve para indicar cómo termina la impresión. Por defecto termina con salto de línea (`\n`).


````python
print("Hola", end=" ")
print("mundo")

print("Línea 1", end="***")
print("Línea 2")
````

Estos dos parámetros aparecen con frecuencia en ejercicios básicos y conviene dominarlos bien.


## 6. Formas de construir salidas de texto

Hay varias formas habituales de mostrar información.

### 6.1 Concatenación


````python
nombre = "Luis"
print("Hola " + nombre)
````

La concatenación funciona bien con cadenas, pero obliga a convertir manualmente si intervienen números.


````python
edad = 25
print("Edad: " + str(edad))
````

### 6.2 Separando con comas en `print()`


````python
edad = 25
print("Edad:", edad)
````

### 6.3 f-strings

Es una de las formas más cómodas y legibles.


````python
nombre = "Marta"
edad = 22
nota = 8.75

print(f"Nombre: {nombre}")
print(f"Edad: {edad}")
print(f"Nota: {nota}")
````

En la práctica docente conviene insistir bastante en las **f-strings**, porque son muy útiles y claras.


## 7. Secuencias de escape importantes

Las secuencias de escape permiten representar caracteres especiales dentro de una cadena.

Las más importantes para PCAP suelen ser:

- `\\n` → salto de línea
- `\\t` → tabulación
- `\\\\` → barra invertida
- `\\"` → comilla doble
- `\\'` → comilla simple


````python
print("Primera línea\nSegunda línea")
print("Columna1\tColumna2\tColumna3")
print("Ruta: C:\\Users\\Jose")
print("Ella dijo: \"Hola\"")
print('It\'s correct')
````

## 8. Introducción a archivos

Además de trabajar con teclado y pantalla, los programas pueden leer y escribir archivos.

Para ello se usa normalmente la función `open()`.

### Sintaxis general

```python
open(nombre_archivo, modo)
```

- `nombre_archivo`: archivo que se quiere abrir
- `modo`: cómo se quiere abrir ese archivo


## 9. Modos de apertura más importantes

Los modos principales que conviene conocer para PCAP son:

- `"r"` → lectura
- `"w"` → escritura
- `"a"` → añadir al final
- `"x"` → crear archivo nuevo
- `"t"` → modo texto
- `"b"` → modo binario

### Resumen rápido

- `"r"` falla si el archivo no existe.
- `"w"` crea el archivo si no existe y **borra el contenido anterior** si existe.
- `"a"` crea el archivo si no existe y añade contenido al final.
- `"x"` crea el archivo, pero falla si ya existe.


## 10. Escritura básica con archivos

### Modo `"w"`


````python
with open("ejemplo_io_1.txt", "w", encoding="utf-8") as archivo:
    archivo.write("Primera línea\n")
    archivo.write("Segunda línea\n")

print("Archivo escrito correctamente.")
````

### Punto importante

`write()` escribe texto y devuelve el número de caracteres escritos.


````python
with open("ejemplo_io_2.txt", "w", encoding="utf-8") as archivo:
    cantidad = archivo.write("Hola archivo")
    print("Caracteres escritos:", cantidad)
````

## 11. Lectura básica con archivos

### 11.1 `read()`

Lee todo el contenido del archivo como una sola cadena.


````python
with open("ejemplo_io_1.txt", "r", encoding="utf-8") as archivo:
    contenido = archivo.read()

print(contenido)
print(type(contenido))
````

### 11.2 `readline()`

Lee una línea cada vez.


````python
with open("ejemplo_io_1.txt", "r", encoding="utf-8") as archivo:
    linea1 = archivo.readline()
    linea2 = archivo.readline()

print("Línea 1:", linea1)
print("Línea 2:", linea2)
````

### 11.3 `readlines()`

Devuelve una lista de líneas.


````python
with open("ejemplo_io_1.txt", "r", encoding="utf-8") as archivo:
    lineas = archivo.readlines()

print(lineas)
print(type(lineas))
````

## 12. Recorrer un archivo línea a línea

Una forma muy habitual y recomendable es recorrer el archivo con un bucle `for`.


````python
with open("ejemplo_io_1.txt", "r", encoding="utf-8") as archivo:
    for linea in archivo:
        print(">>", linea.strip())
````

Esto suele ser más práctico que usar `readlines()` cuando se quieren procesar líneas una a una.


## 13. Añadir contenido con modo `"a"`

El modo `"a"` añade información al final del archivo.


````python
with open("ejemplo_io_1.txt", "a", encoding="utf-8") as archivo:
    archivo.write("Tercera línea añadida\n")

with open("ejemplo_io_1.txt", "r", encoding="utf-8") as archivo:
    print(archivo.read())
````

## 14. Crear archivo nuevo con modo `"x"`

El modo `"x"` crea un archivo nuevo, pero da error si ya existe.


````python
try:
    with open("archivo_nuevo_x.txt", "x", encoding="utf-8") as archivo:
        archivo.write("Archivo creado con modo x")
    print("Archivo creado correctamente.")
except FileExistsError as e:
    print("Error:", e)
````

## 15. Diferencia entre texto y binario

### Modo texto (`t`)
Es el modo habitual cuando se trabaja con cadenas.

### Modo binario (`b`)
Se usa para datos binarios, como imágenes, audio o archivos no textuales.

Para PCAP suele bastar con entender la diferencia conceptual.


````python
with open("binario_ejemplo.dat", "wb") as archivo:
    archivo.write(b"ABC123")

with open("binario_ejemplo.dat", "rb") as archivo:
    datos = archivo.read()

print(datos)
print(type(datos))
````

En modo binario se trabaja normalmente con objetos de tipo `bytes`, no con `str`.


## 16. Cerrar archivos con `close()`

Tradicionalmente se abrían archivos así:


````python
archivo = open("ejemplo_close.txt", "w", encoding="utf-8")
archivo.write("Texto de prueba")
archivo.close()

print("Archivo cerrado manualmente.")
````

Esto funciona, pero hoy en día se recomienda usar `with open(...)` porque cierra el archivo automáticamente incluso si ocurre un error.


## 17. Uso recomendado: `with open(...) as ...`

Este es uno de los puntos más importantes para PCAP y para programar bien en general.


````python
with open("ejemplo_with.txt", "w", encoding="utf-8") as archivo:
    archivo.write("Uso correcto de with")

print("El archivo ya ha sido cerrado automáticamente.")
````

### Ventajas de `with`

- El archivo se cierra automáticamente.
- El código es más seguro.
- Es la forma más limpia y recomendada.


## 18. `writelines()`

Permite escribir varias líneas de una vez a partir de una colección de cadenas.


````python
lineas = ["Uno\n", "Dos\n", "Tres\n"]

with open("ejemplo_writelines.txt", "w", encoding="utf-8") as archivo:
    archivo.writelines(lineas)

with open("ejemplo_writelines.txt", "r", encoding="utf-8") as archivo:
    print(archivo.read())
````

### Observación importante

`writelines()` **no añade automáticamente saltos de línea**. Si se quieren líneas separadas, hay que incluir `\n` manualmente.


## 19. Posición del puntero: `tell()` y `seek()`

Cuando se trabaja con archivos, existe una posición interna llamada **puntero** que indica por dónde va la lectura o escritura.


````python
with open("ejemplo_puntero.txt", "w", encoding="utf-8") as archivo:
    archivo.write("abcdefg")

with open("ejemplo_puntero.txt", "r", encoding="utf-8") as archivo:
    print("Posición inicial:", archivo.tell())

    print("Leo 3 caracteres:", archivo.read(3))
    print("Posición actual:", archivo.tell())

    archivo.seek(0)
    print("Posición tras seek(0):", archivo.tell())
    print("Vuelvo a leer 2 caracteres:", archivo.read(2))
````

### Idea importante

- `tell()` indica la posición actual.
- `seek(pos)` mueve el puntero a una posición concreta.


## 20. Errores comunes al trabajar con archivos

### 20.1 `FileNotFoundError`

Ocurre cuando se intenta abrir para lectura un archivo que no existe.


````python
try:
    with open("archivo_que_no_existe.txt", "r", encoding="utf-8") as archivo:
        print(archivo.read())
except FileNotFoundError as e:
    print("Error:", e)
````

### 20.2 `PermissionError`

Puede ocurrir cuando no se tienen permisos adecuados sobre el archivo o la ruta.

No siempre es fácil reproducirlo de forma segura en un entorno de ejemplo, pero conviene conocerlo conceptualmente.


### 20.3 Conversión de tipos no válida

También es frecuente que el usuario escriba un valor que no puede convertirse a entero o flotante.


````python
try:
    numero = int("hola")
except ValueError as e:
    print("Error:", e)
````

## 21. Manejo básico de errores con `try-except`

Cuando se trabaja con I/O, es muy habitual proteger el código con `try-except`.


````python
try:
    with open("ejemplo_io_1.txt", "r", encoding="utf-8") as archivo:
        contenido = archivo.read()
    print(contenido)
except FileNotFoundError:
    print("El archivo no existe.")
except PermissionError:
    print("No hay permisos suficientes.")
except Exception as e:
    print("Se ha producido otro error:", e)
````

Para PCAP no hace falta complicarlo demasiado, pero sí conviene que quede clara la idea de capturar errores habituales.


## 22. Conceptos básicos: stdin, stdout y stderr

A nivel conceptual conviene conocer estos términos:

- **stdin**: entrada estándar
- **stdout**: salida estándar
- **stderr**: salida de error estándar

### Explicación simple

- `input()` normalmente lee desde **stdin**.
- `print()` normalmente escribe en **stdout**.
- Los mensajes de error suelen ir a **stderr**.

Para PCAP suele bastar con una comprensión básica de estos conceptos.


## 23. Resumen rápido para PCAP

### Entrada y salida por consola
- `input()` siempre devuelve `str`.
- Normalmente hay que convertir con `int()` o `float()`.
- `print()` permite usar `sep` y `end`.
- Las f-strings son una forma muy recomendable de mostrar datos.

### Archivos
- Se abren con `open()`.
- Modos importantes: `"r"`, `"w"`, `"a"`, `"x"`.
- Lectura: `read()`, `readline()`, `readlines()`.
- Escritura: `write()`, `writelines()`.
- Es preferible usar `with open(...)`.
- `tell()` y `seek()` controlan la posición del puntero.
- Errores importantes: `FileNotFoundError`, `PermissionError`.

### Ideas clave
- La entrada por teclado llega como texto.
- La salida puede ir a pantalla o a archivos.
- Los archivos deben cerrarse correctamente.
- `with` es la forma más segura y recomendada.

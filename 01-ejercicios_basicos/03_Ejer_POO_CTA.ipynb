## 11) Ejercicios propuestos
1) Crea una clase `Coche` con atributos `marca`, `modelo`, `km` y un método `conducir(km)` que aumente los kilómetros.

2) Añade un atributo “privado” `__vin` (número de bastidor) y un método `get_vin()`.

3) Crea una función `inspeccionar(obj)` que imprima:
- si tiene atributo `km` (`hasattr`)
- su `__dict__` (si existe)

4) Implementa un método `destroy()` que marque el coche como `activo=False`.


```python
class Coche:
    def __init__(self, marca, modelo, km ,vin):
        self.marca = marca
        self.modelo = modelo
        self.km = km
        self.__vin = None # 2.Creacion atributo privado
        
    def condudir(self,km):
        self.km += km
        
    def get_vin(self):
        return self.__vin
    
    # 4. Destroy
    def destroy(self):
        self.activo = False
```

```python
# 3. Funcion inspeccion attr y dic
def inspeccionar(obj):
    print("Tiene km?", hasattr(obj, "km"))
    
    if hasattr(obj, "__dict__"):
        print("Atributos (__dict__):", obj.__dict__)
    else:
        print("No tiene __dict__")
```

```python
# ---------------------------
# PRUEBAS DE EJECUCIÓN
# ---------------------------

coche1 = Coche("Toyota", "Corolla", 12000, "VIN12345")

print("Marca:", coche1.marca)
print("Modelo:", coche1.modelo)
print("KM iniciales:", coche1.km)

# Conducir
coche1.conducir(150)
print("KM después de conducir:", coche1.km)

# Obtener VIN
print("VIN:", coche1.get_vin())

# Inspeccionar objeto
inspeccionar(coche1)

# Destroy
coche1.destroy()
print("¿Coche activo?:", coche1.activo)
```

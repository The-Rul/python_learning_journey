# Solución — POO Python (PCAP): Clase `Coche`, atributo privado, `hasattr`, `__dict__` y `destroy()`

**Enunciado**  
1) Crea una clase `Coche` con atributos `marca`, `modelo`, `km` y un método `conducir(km)` que aumente los kilómetros.  
2) Añade un atributo “privado” `__vin` (número de bastidor) y un método `get_vin()`.  
3) Crea una función `inspeccionar(obj)` que imprima:
   - si tiene atributo `km` (`hasattr`)
   - su `__dict__` (si existe)  
4) Implementa un método `destroy()` que marque el coche como `activo=False`.

> Nota: En Python, `__vin` usa **name mangling** (se transforma internamente a `_Coche__vin`).


## 1) Implementación de la clase `Coche`

```
class Coche:
    def __init__(self, marca: str, modelo: str, km: float, vin: str):
        self.marca = marca
        self.modelo = modelo
        self.km = float(km)
        self.activo = True              # estado del coche (por defecto activo)
        self.__vin = vin                # atributo "privado" (name mangling)

    def conducir(self, km: float) -> None:
        """Aumenta los kilómetros recorridos.

        - Si el coche está inactivo (destroy), no permite conducir.
        - No permite sumar kilómetros negativos.
        """
        if not self.activo:
            raise RuntimeError("El coche está inactivo. No se puede conducir.")
        km = float(km)
        if km < 0:
            raise ValueError("Los kilómetros a conducir no pueden ser negativos.")
        self.km += km

    def get_vin(self) -> str:
        """Devuelve el número de bastidor."""
        return self.__vin

    def destroy(self) -> None:
        """Marca el coche como inactivo."""
        self.activo = False
```

## 2) Función `inspeccionar(obj)` usando `hasattr` y `__dict__`

```
def inspeccionar(obj) -> None:
    # 1) Comprobar si existe el atributo 'km'
    print("¿Tiene atributo 'km'? ->", hasattr(obj, "km"))

    # 2) Mostrar __dict__ si existe (no todos los objetos lo tienen)
    if hasattr(obj, "__dict__"):
        print("__dict__ ->", obj.__dict__)
    else:
        print("No tiene __dict__ (por ejemplo, si usa __slots__ o es un tipo built-in).")
```

## 3) Prueba rápida de funcionamiento (demo)

```
c = Coche(marca="Toyota", modelo="Corolla", km=12000, vin="JTDBR32E123456789")

print("Marca:", c.marca)
print("Modelo:", c.modelo)
print("KM inicial:", c.km)
print("Activo:", c.activo)

# Conducir 150 km
c.conducir(150)
print("KM tras conducir 150:", c.km)

# Acceso al VIN mediante método (forma correcta)
print("VIN (get_vin):", c.get_vin())

# Inspección del objeto
inspeccionar(c)
```

## 4) Verificación de `destroy()`

`destroy()` marca el coche como `activo=False`. En esta implementación, además, `conducir()` lanza un error si el coche no está activo.


```
c.destroy()
print("Activo tras destroy():", c.activo)

try:
    c.conducir(10)
except RuntimeError as e:
    print("Error al conducir con coche inactivo:", e)
```

## 5) (Opcional) Demostración de *name mangling* del atributo `__vin`

El atributo `__vin` no se accede directamente como `c.__vin`.  
Internamente se renombra a `_Coche__vin`.


```
# Esto fallará:
try:
    print(c.__vin)
except AttributeError as e:
    print("Acceso directo a c.__vin ->", e)

# Pero existe con el nombre "mangleado":
print("Acceso con name mangling:", c._Coche__vin)
```
